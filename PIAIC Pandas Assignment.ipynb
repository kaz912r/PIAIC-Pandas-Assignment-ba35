```python
%matplotlib inline
import pandas as pd 
import numpy as np
import regex as re
import matplotlib.pyplot as plt
df1= pd.read_csv(r'C:\Users\User1\Desktop\Assignment\states0.csv')
df2= pd.read_csv(r'C:\Users\User1\Desktop\Assignment\states1.csv')
df3= pd.read_csv(r'C:\Users\User1\Desktop\Assignment\states2.csv')
df4= pd.read_csv(r'C:\Users\User1\Desktop\Assignment\states3.csv')
df5= pd.read_csv(r'C:\Users\User1\Desktop\Assignment\states4.csv')
df6= pd.read_csv(r'C:\Users\User1\Desktop\Assignment\states5.csv')
```

```python
us_census['Income($)']= us_census['Income'].str.replace('\$','',regex=True)
us_census.drop(['Unnamed: 0','Income'], axis=1, inplace=True)
```

```python
us_census[['Men','Women']]=us_census['GenderPop'].str.split('_',expand=True)
```

```python
us_census['Women']= us_census['Women'].str.replace('F$','',regex=True)
us_census['Men']= us_census['Men'].str.replace('M$','',regex=True)
```

```python
us_census['Men'] = us_census['Men'].astype(int) 
```

```python
us_census['Women'].replace('',0,regex=True,inplace=True)
```

```python
us_census['Women']=us_census['Women'].astype(int)
# us_census['Women']
```

```python
plt.scatter(us_census['Women'],us_census['Income($)'])
plt.xticks(rotation=40)
plt.grid()
plt.xlabel('Women Population')
plt.ylabel('Income($)')
plt.title('Women vs Income')
plt.figure(figsize=(10.5,18.5))
plt.show()
```

```python
us_census['Asian']= us_census['Asian'].str.replace('\%','',regex=True)
us_census['Black']= us_census['Black'].str.replace('\%','',regex=True)
us_census['Hispanic']= us_census['Hispanic'].str.replace('\%','',regex=True)
us_census['Native']= us_census['Native'].str.replace('\%','',regex=True)
us_census['White']= us_census['White'].str.replace('\%','',regex=True)
us_census['Pacific']= us_census['Pacific'].str.replace('\%','',regex=True)
```
